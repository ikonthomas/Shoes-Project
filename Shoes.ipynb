**Added Columns**

We created 3  columns.

Column **'Price.stat'** contains boolean value 'True' if the shoe price is over 50 or 'False' if the price is under 50

Column **'brand.count'** is the number of shoes that brand sold

Column **'price_band'** has values 0,1,2. The price of the shoe is <=20 column value 0, >20 and <=50 column value 1 and >50 column value 2.

```
#creating a new column with boolean values.Shoes priced over $50 is expensive,under 50 is affordable.
shoes_USD['price.stat']=(shoes_USD['prices.amountmax']>50)
shoes_USD.head().T
```

```
# New column brand count . columnvalue says how many pair of shoes the brand sold
shoes_USD['brand.count'] = shoes_USD['brand'].map(shoes_USD['brand'].value_counts())
shoes_USD[['brand','brand.count']]
```

```
affordable=shoes_USD.sort_values(by='prices.amountmax')
affordable=affordable[['brand','prices.amountmax']].groupby('brand').agg(min)<50 #.filter(filter_max)
affordable.head(10)
```

```
brand_price=shoes_USD.groupby('brand').aggregate(['min', np.median, max])
brand_price.head().T
```

```

# creating price band to group the prices in data. price>=20 = 0,price>30 & <=50 =1, price>50 =2
shoes_USD['price_band']=0
shoes_USD.loc[shoes_USD['prices.amountmax']<=20,'price_band']=0
shoes_USD.loc[(shoes_USD['prices.amountmax']>20)&(shoes_USD['prices.amountmax']<=50),'price_band']=1
shoes_USD.loc[shoes_USD['prices.amountmax']>50,'price_band']=2
shoes_USD.head(10).T
```

```
# checking the value counts in price_band
shoes_USD['price_band'].value_counts()
```

```
# Histogram of price range( <=30 is 0, >30 and <=100 is 1 and >100 is 2)
shoes_USD['price_band'].plot(kind='hist')
```

# Label encoding

The next step in data preparation is label encoding. Our data set contain many categorical variables and the values are stored in text format. To convert these values text to numbers we use a technique called label encoding. Here we change the 'brand','colors','prices.merchant' columns values from words to numbers.

The encoded values are stored in to column names **'brand_code','colors_code','merchant_code'**.

Also changed the boolean values to 0 and 1 by simply changing their type to integer.

```

# checking datatypes 
shoes_USD.info()
```

```
# Converting the boolean values to 0,1 using .astype()
shoes_USD[['prices.issale','price.stat']] = shoes_USD[['prices.issale','price.stat']].astype('int')
shoes_USD.dtypes
```

```
shoes_USD[['prices.issale','price.stat']].head()
```

```
shoes_USD['price.stat'].value_counts()
```

```
#Encode the brand values which are strings to integers.
from sklearn.preprocessing import LabelEncoder

lb_brand = LabelEncoder()
shoes_USD['brand_code'] = lb_brand.fit_transform(shoes_USD['brand'])
shoes_USD[['brand', 'brand_code']].head(10)
```

```
lb_brand = LabelEncoder()
shoes_USD['colors_code'] = lb_brand.fit_transform(shoes_USD['colors'])
shoes_USD[['colors', 'colors_code']].head(10)
```

```
lb_brand = LabelEncoder()
shoes_USD['merchant_code'] = lb_brand.fit_transform(shoes_USD['prices.merchant'])
shoes_USD[['prices.merchant', 'merchant_code']].head(10)
```
